```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import nltk
import pandas as pd
import re
```

```python
paragraph="""Beans. I was trying to explain to somebody as we were flying in, that’s corn. That’s beans. And they were very impressed at my agricultural knowledge. Please give it up for Amaury once again for that outstanding introduction. I have a bunch of good friends here today, including somebody who I served with, who is one of the finest senators in the country, and we’re lucky to have him, your Senator, Dick Durbin is here. I also noticed, by the way, former Governor Edgar here, who I haven’t seen in a long time, and somehow he has not aged and I have. And it’s great to see you, Governor. I want to thank President Killeen and everybody at the U of I System for making it possible for me to be here today. And I am deeply honored at the Paul Douglas Award that is being given to me. He is somebody who set the path for so much outstanding public service here in Illinois. Now, I want to start by addressing the elephant in the room. I know people are still wondering why I didn’t speak at the commencement."""
```

```python
nltk.download('stopwords')
```

```python
stemmer=PorterStemmer()
lemmatizer=WordNetLemmatizer()
stop_words=set(stopwords.words('english'))
```

```python
nltk.download('punkt')
```

```python
Sentence=nltk.sent_tokenize(paragraph)
Sentence
```

```python
for x in Sentence:
  print(x)
```

```python
corpus=[]
for i in range(len(Sentence)):
  #remove all other char except a-z and A-Z
  sent=re.sub('[^a-zA-z]',' ',Sentence[i])

  #Convert into Lower Casing
  sent=sent.lower()

  #Steamming
  sent=" ".join([stemmer.stem(word) for word in str(sent).split()])

  #Remove Stopword
  sent=" ".join([word for word in str(sent).split() if(word not in stop_words)])
  corpus.append(sent)
corpus
```

```python
#Stem+BOW
from sklearn.feature_extraction.text import TfidfVectorizer
cv=TfidfVectorizer(max_features=30)
X=cv.fit_transform(corpus).toarray()
X
```
