#Expand Contraction 

```python
import contractions 
text ="I'll be there in 5min"
print("Before contractions Text:- ",text)

text=contractions.fix(text)
print("After contractions Text :- ",text)
```

#Lower Casing

```python
text ="I'll be there in 5min"
print("Before Lower Casing:- ",text)

text=contractions.fix(text).lower()
print("After Lower Casing :- ",text)
```

#Remove Punctuation

```python
import string
punctuation=string.punctuation
text ="I'll be there in 5min."
print("Before Punctuation:- ",text)
for punc in punctuation:
  text=text.replace(punc," ")
print("After Punctuation :- ",text)

```

# Remove Stopword

```python
# import nltk
# nltk.download('stopwords')
```

```python
from nltk.corpus import stopwords
stopword=set(stopwords.words('english'))

text ="I'll be there in 5min"
print("Before stopword Text:- ",text)
temp=[]
for word in str(text).split():
  if(word not in stopword):
    temp.append(word)
text=" ".join(temp)
print("After stopwords text :- ",text)
```

```python
from gensim.parsing.preprocessing import remove_stopwords
text ="I'll be there in 5min"
print("Before contraction Text :- ",text)
remove_stopwords(text)
```

# Stemming

```python
# it is use to convert the word into base form
# running=run
```

```python
from nltk.stem import PorterStemmer
text="I like python Programing more than anything"
print("before stem text:- ",text)
port_stem=PorterStemmer()
def stem_word(text):
  temp=" ".join([port_stem.stem(word) for word in str(text).split()])
  return temp
print("After Stem Text:- ",stem_word(text))
```

```python
stem_word("I likes History as it is very interisting topic")
```

```python
from nltk.stem import WordNetLemmatizer
text="I like python Programing more than anything"
print("before lemmatize text:- ",text)
lemmatize_obj=WordNetLemmatizer()
def lemmatize_word(text):
  temp=" ".join([lemmatize_obj.lemmatize(word) for word in str(text).split()])
  return temp
print("After lemmatize Text:- ",stem_word(text))
```

```python
lemmatize_word("I likes History as it is very interisting topic")
```
