```python
# split data into x and y
#X=mass,width,height,color_score
#y=fruit_name
```

```python
X=df[['mass', 'width', 'height','color_score']]
y=df['fruit_name']
```

```python
X.head()
```

```python
y.head()
```

```python
# split data into traning and testing
# train_test_split
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=3)
```

```python
print("X_train:-",X_train.shape)
print("X_test:-",X_test.shape)
print("y_train:-",y_train.shape)
print("y_test:-",y_test.shape)
```

# Step 5:Building a machine learning model

```python
# create a model k=3 ,distance= manhatten_distance
ML_model=KNeighborsClassifier(n_neighbors=3,p=1)

# model traning
ML_model.fit(X_train,y_train)

# find y_pred
y_pred=ML_model.predict(X_test)

# find accuracy
acc=accuracy_score(y_test,y_pred)
print("acc for k=3",acc*100)
```

```python
# create a model k=3 ,distance= Eucliden_distance
ML_model=KNeighborsClassifier(n_neighbors=3,p=2)

# model traning
ML_model.fit(X_train,y_train)

# find y_pred
y_pred=ML_model.predict(X_test)

# find accuracy
acc=accuracy_score(y_test,y_pred)
print("acc for k=3",acc*100)
```

# step 6: Model Evaluation & Optimization finding Kvalue in KNN

```python
# creat model k=1,3,5....29
accuracy=[]

for k in range(1,30,2):
    #create model
    ML_model=KNeighborsClassifier(n_neighbors=k)
    #Model train
    ML_model.fit(X_train,y_train)
    #find y_pred
    y_pred=ML_model.predict(X_test)
    #find accuracy
    acc=accuracy_score(y_test,y_pred)
    accuracy.append(acc)
```

```python
accuracy
```

```python
plt.plot(list(range(1,30,2)),accuracy,'-*')
plt.xlabel("K-values")
plt.ylabel("Accuracy")
```

```python
# from above graph k=5 having good accuracy score
# so k=5 isbest values for fruit dataset
```

```python
# create a model k=5 ,distance= manhatten_distance
ML_model=KNeighborsClassifier(n_neighbors=5,p=1)

# model traning
ML_model.fit(X_train,y_train)

# find y_pred
y_pred=ML_model.predict(X_test)

# find accuracy
acc=accuracy_score(y_test,y_pred)
print("acc for k=5",acc*100)
```

# Model Eval Parameter

```python
# Accuracy and ER

acc=accuracy_score(y_test,y_pred)
print("Accuracy :-",acc)
print("Error rate :-",1-acc)
```

```python
#Confusion matrix

from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(10,5))

sns.heatmap(cm,annot=True)
plt.xlabel('Predicted')
plt.ylabel("Actual")
```

```python
from sklearn.metrics import classification_report
lab=['mass', 'width', 'height','color_score']
print(classification_report(y_test,y_pred,target_names=lab))
```

```python
from sklearn.metrics import precision_score,recall_score,f1_score
acc=accuracy_score(y_test,y_pred)
print("Accuracy :-",acc)
print("Error rate :-",1-acc)
print("Precision:-",precision_score(y_test,y_pred,average='weighted'))
print("Recall:- ",recall_score(y_test,y_pred,average='weighted'))
print("F1 score :- ",f1_score(y_test,y_pred,average='weighted'))

#Confusion matrix
from sklearn.metrics import confusion_matrix
cm=confusion_matrix(y_test,y_pred)
print(cm)
plt.figure(figsize=(10,5))

sns.heatmap(cm,annot=True,cmap="YlGnBu")
plt.xlabel('Predicted')
plt.ylabel("Actual")

```
