```python
import pandas as pd
from sklearn.feature_selection import  VarianceThreshold
```

```python
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Dataset/Santander Customer Satisfaction/Santander Customer Satisfaction.csv')
df.head()
```

```python
df.info()
```

```python
X=df.drop(labels='TARGET',axis=1)
y=df['TARGET']
```

```python
# split the data 
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.3,random_state=3)
```

```python
var_threshold=VarianceThreshold(threshold=0)
var_threshold.fit(X_train)
```

```python
sum(var_threshold.get_support())
```

```python
zero_var_col=[]
for columns in X_train.columns:
  if(columns not in X_train.columns[var_threshold.get_support()]):

    zero_var_col.append(columns)
```

```python
print(zero_var_col)
```

```python
print(len(zero_var_col))
```

```python
# Before removing cols with var=0

print("X_train cols len:-",len(X_train.columns))
print("X_test cols len:-",len(X_test.columns))
```

```python
# Drop cols with var=0
X_train.drop(labels=zero_var_col,axis=1,inplace=True)
X_test.drop(labels=zero_var_col,axis=1,inplace=True)
```

```python
# After removing cols with var=0

print("X_train cols len:-",len(X_train.columns))
print("X_test cols len:-",len(X_test.columns))
```
